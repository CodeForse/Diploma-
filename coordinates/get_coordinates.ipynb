```python
def get_adress(row):
    if 'этаж' not in row['title'].lower():
        adress = row['title'][row['title'].find('м²') + 3:]
    else:
        adress =row['title'][row['title'].find('этаж') + 5:]
    
    if 'р-н' not in adress:
        adress = row['city'].replace('Алматы', '').replace(',','') + ', '*(row['city']!='Алматы')+ adress.strip()
    return adress
```

```python
df = pd.read_excel('./data/constructed/all_addresses.xlsx')
```

```python
df = df.drop(columns = ['Unnamed: 0'])
```

```python
df = df.rename(columns={0: 'address'}).drop(columns = ['Unnamed: 0'])
```

```python
def get_coordinates(x):
    try:
        #2gis
        api_url = f'https://catalog.api.2gis.com/3.0/items/geocode?q={x}&page_size=1&fields=items.point&key=rugpth3331'
        #opengatedata
        # https://api.opencagedata.com/geocode/v1/json?q=URI-ENCODED-PLACENAME&key=1f36bc68ad414f18ab8369b1066d0598
        r = requests.get(api_url)
        lat = r.json()['result']['items'][0]['point']['lat']
        long = r.json()['result']['items'][0]['point']['lon']
        return lat, long
    except:
        return np.NaN, np.NaN
```

```python
df[['Lattitude', 'Longitude']] = df['address'].apply(lambda x: pd.Series(get_coordinates(x)))
```
