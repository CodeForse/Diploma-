```python
def make_pattern_floor(x: str):
    try:
        vals = re.findall(r'\d+//\d+', x)[0].split('//')
        return f'{vals[1]} из {vals[0]}'
    except IndexError:
        return -1
```

```python
df_2021txt.loc[df_2021txt['flat.floor'].isna(), 'flat.floor'] = df_2021txt[df_2021txt['flat.floor'].isna()]['title'].apply(make_pattern_floor)
```
